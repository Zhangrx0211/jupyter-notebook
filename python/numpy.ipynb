```python
import numpy as np

#创建整数数组
arr1 = np.array([1, 2, 3, 4])
print(arr1) 

#创建浮点数数组
arr2 = np.array([1.0, 2.0, 3.0, 4.0])
print(arr2) 
```

```python
#给arr1插入浮点数,被截断，数组仍为整数
arr1[0] = 20.5
print(arr1) 
```

```python
#整数数组转换为浮点数数组
arr1_float = arr1.astype(float)
print(arr1_float)

#浮点数数组转换为整数数组
arr2_int = arr2.astype(int)
print(arr2_int)
```

```python
#一维数组
a=np.array([1, 2, 3])
print(a)
print(a.ndim)  # 输出维度
print(a.shape)  # 输出形状
print(a.size)   # 输出元素数量

#二维数组（两个中括号）
b=np.array([[4, 5, 6]])
print(b)
print(b.ndim)  # 输出维度
print(b.shape)  # 输出形状,查看几行几列，这里是1行3列
print(b.size)   # 输出元素数量
```

```python
#索引访问元素
print(a[0])  # 输出1

#赋值
a[0] = 10
print(a)  

#切片
print(a[1:3])  
```

```python
#二维数组索引
c=np.array([[4, 5, 6],[7, 8, 9]])
print(c)  # 输出二维数组
print(c[0, 1])  # 逗号前面是行，逗号后面是列，输出5
print(c[1, 2])  
print(c[0, :])  # 输出第一行
print(c[:, 1])  # 输出第二列
```

```python
#创建基础数组
np.zeros(5)  # 创建一个包含5个元素的全零数组
```

```python
np.ones(5)   # 创建一个包含5个元素的全一数组
```

```python
np.arange(0, 10, 2)  # 创建一个从0到10（不包括10），步长为2的数组
```

```python
np.linspace(0, 1, 5)  # 创建一个从0到1的等差数列(每两个数间距一样)，包含5个元素
```

```python
np.ones(5, dtype=int)  # 创建一个包含5个元素的全一数组(整数类型)
```

```python
#排序
arr = np.array([3, 1, 4, 1, 5])
a=np.sort(arr) #保存排序后的数组
a
```
