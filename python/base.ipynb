1基本变量类型

```python
#1.1字符串
str1="Hello,"
str2=" World!"
str3=f"{str1}{str2}" #字符串拼接
print(str3) #输出: Hello, World!  

answer=42
print(f"结果是 {answer}") #输出: 结果是 42

#\n表示换行，\t表示制表符
message="shop:\n apple,\n\t banana,\n orange."
print(message) 
```

```python
#1.2数字
a = 2**4 #幂运算，输出: 16
b = 28 // 5 #整数除法，输出: 5
c = 28 % 5 #取模运算，输出: 3
print(a) 
print(b) 
print(c) 
```

```python
#1.3布尔值 True和False是Python中的布尔值，常用于条件判断和逻辑运算。
str1="Hello"
str2="World"
print(str1 == str2) #输出: False
print(str1 != str2) #输出: True

set1={1, 2, 3} #集合，元组，列表和字典都可以
print(2 in set1) #输出: True
print(4 in set1) #输出: False

#and和or是Python中的逻辑运算符，用于连接布尔表达式。
x = 5
print(x > 0 and x < 10) #全部条件都满足，则为True
print(x > 0 or x > 10) #至少一个条件满足，则为True
```

```python
#1.4基本变量转换，字符串、整数、浮点数和布尔值之间的转换。
num_str = "42"
num_int = 123
num_float = 3.14
bool_val = True

#转换为字符串，其他类似
print(str(num_int)) 
print(str(num_float)) 
print(str(bool_val))  
```

3类：在一堆函数之间传递参数

```python
#3.1类中的函数：方法。一个类包含一个__init__方法和很多自定义方法
class Counter:
    def __init__(self,a,b): #特殊方法
        #a,b公共变量,也是self的属性
        self.a = a          
        self.b = b

    def add(self): #自定义方法,➕
        return self.a + self.b
    def mul(self):
        return self.a * self.b
```

```python
#创建类的实例
counter = Counter(3, 4)
print(counter.a) #访问实例的属性，输出: 3
print(counter.add()) #调用add方法，输出: 7
print(counter.mul()) #调用mul方法，输出: 12
```

```python
#3.2继承
#父类
class Counter:
    def __init__(self,a,b): #特殊方法
        #a,b公共变量,也是self的属性
        self.a = a          
        self.b = b

    def add(self): #自定义方法,➕
        return self.a + self.b
#子类
class Counter2(Counter): #Counter2继承Counter
    def __init__(self,a,b): 
        #引用父类的属性和方法
        super().__init__(a,b) #继承父类的属性和方法

    def mul(self):
        return self.a * self.b    

test = Counter2(5, 6)
print(test.a) #访问父类的属性，输出: 5
print(test.add()) #调用父类的方法，输出: 11
print(test.mul()) #调用子类重写的方法，输出: 30
```
