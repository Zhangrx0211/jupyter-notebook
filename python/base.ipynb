1基本变量类型

```python
#1.1字符串
str1="Hello,"
str2=" World!"
str3=f"{str1}{str2}" #字符串拼接
print(str3) #输出: Hello, World!  

answer=42
print(f"结果是 {answer}") #输出: 结果是 42

#\n表示换行，\t表示制表符
message="shop:\n apple,\n\t banana,\n orange."
print(message) 
```

```python
#1.2数字
a = 2**4 #幂运算，输出: 16
b = 28 // 5 #整数除法，输出: 5
c = 28 % 5 #取模运算，输出: 3
print(a) 
print(b) 
print(c) 
```

```python
#1.3布尔值 True和False是Python中的布尔值，常用于条件判断和逻辑运算。
str1="Hello"
str2="World"
print(str1 == str2) #输出: False
print(str1 != str2) #输出: True

set1={1, 2, 3} #集合，元组，列表和字典都可以
print(2 in set1) #输出: True
print(4 in set1) #输出: False

#and和or是Python中的逻辑运算符，用于连接布尔表达式。
x = 5
print(x > 0 and x < 10) #全部条件都满足，则为True
print(x > 0 or x > 10) #至少一个条件满足，则为True
```

```python
#1.4基本变量转换，字符串、整数、浮点数和布尔值之间的转换。
num_str = "42"
num_int = 123
num_float = 3.14
bool_val = True

#转换为字符串，其他类似
print(str(num_int)) 
print(str(num_float)) 
print(str(bool_val))  
```

#2.2元组

```python
#创建元组
(1, 2, 3) #元组使用圆括号()定义。
```

```python
4, 5, 6 #也可以直接用逗号分隔的方式创建元组。
```

```python
#可以容纳所有的数据类型
"a", 1, 3.14, True, [1, 2], (3, 4), {"key": "value"} #元组可以包含不同类型的元素。
```

```python
answer =42
print(f"结果是 {answer}") #f字符串
print("最终是：", answer) #元组法
```

```python
#元组拆分法
a, b, c = 1, 2, 3 #将元组中的元素分别赋值给变量a、b和c
print(b,c,a) 

#元组交换法
c, d = 1, 2
c, d = d, c #交换c和d的值
print(c, d)

#只需要部分元素
values = 1, 2, 3, 4, 5
e, f, *rest = values #使用*rest来捕获剩余的元素
print(e, f, rest) #输出: 1 2 [3, 4, 5]
```

#2.3列表

```python
#创建列表
list1 = [1, 2, 3,"st"] #列表使用方括号[]定义。
print(list1)
print(list1[0]) #访问列表中的元素，输出: 1
list1.append(4) #添加元素4到列表末尾
print(list1) #输出: [1, 2, 3, 'st', 4]
list1[1] = 20 #修改列表中的元素
print(list1) #输出: [1, 20, 3, 'st', 4] 
```

```python
#切片访问列表中的元素，列表和切片是独立的
list2 = [10, 20, 30, 40, 50]
print(list2[1:4]) 
print(list2[2:-2]) #切除开头和结尾2个元素
print(list2[ : :2]) #步长为2，访问列表中的每隔一个元素
```
