## 机器学习主要类型：监督学习 无监督学习
---

### 监督学习(Supervised Learning) (常用)
    监督学习：预测输入(x)到输出(y) 的映射
#### 1,回归(Regression) 预测值
    回归算法：从无限多个可能的数字中，预测一个数字（预测房价）
#### 2,分类(Classification) 预测状态
    分类算法：预测输出类别集合（预测一张图片是猫还是狗）
----

### 无监督学习(Unsupervised Learning)

#### 1,聚类() 进行分组
    聚类算法：处理无标签数据，并自动将它们分组
