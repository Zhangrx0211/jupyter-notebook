```python
"""
4,实现线性回归模型
    线性回归模型的公式是:f_wb = w * x + b
    x 是输入数据,w 是权重:决定直线的倾斜程度,b 是偏置:决定直线上下移动的位置 
    f_wb 是模型预测出来的结果

    通俗理解：
        这个函数就像一个“计算机器”
        你给它 x、w、b
        它就按照公式 w * x + b
        帮你算出每一个 x 对应的预测结果

        
    np.zeros(m) 表示创建 m 个 0    
"""
# 定义一个函数，用来计算线性回归模型的预测值
# 函数名是 compute_model_output
# x 表示输入数据，w 表示权重，b 表示偏置，是这个函数需要的三个输入
def compute_model_output(x, w, b):

    # 获取训练数据的数量
    # x.shape[0] 表示 x 这个数组中有多少个数据
    m = x.shape[0]  # 比如 x_train = [1.0, 2.0]，那么 m = 2

    # 创建一个全是 0 的数组，用来保存每一个 x 对应的预测结果
    f_wb = np.zeros(m) #如果 m = 2，那么 f_wb 一开始就是 [0.0, 0.0]

    # 使用 for 循环，依次计算每一个 x 的预测值
    # range(m) 表示生成 0 到 m-1 的数字
    for i in range(m):  # 如果 m = 2，那么 i 会依次等于 0、1

        # 使用线性回归公式计算预测值
        # x[i] 表示第 i 个输入数据
        # w * x[i] + b 表示预测结果
        f_wb[i] = w * x[i] + b

    # 返回所有预测结果
    return f_wb
```

## 5. 成本函数 Cost Function (策略) 7-9

### 1. 这一节要解决什么问题？

前面我们已经有了线性回归模型：

$$
\hat{y} = f_{w,b}(x) = wx + b
$$

但是现在还有一个问题：

> 怎么判断一组 $w$ 和 $b$ 好不好？

成本函数的作用就是：

> 用一个数字衡量模型预测得好不好。

---

### 2. 核心概念

成本函数英文叫：**Cost Function**

通常记作：J(w,b),表示：当前这组 $w$ 和 $b$ 对应的预测误差有多大。越小越好

---

### 3. 预测值和真实值

对于一个训练样本：

- 真实值：$y$
- 预测值：$\hat{y}$
- 预测公式：$\hat{y} = wx + b$

误差可以写成：$\hat{y}$ - y

也就是： 预测值 - 真实值

---

### 4. 成本函数公式

线性回归常用的成本函数是：

$$ J(w,b) = \frac{1}{2m}\sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)})^2 $$

也可以写成：

$$ J(w,b) = \frac{1}{2m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)}) - y^{(i)})^2 $$

---

### 5. 公式里的符号是什么意思？

- $J(w,b)$：成本函数
- $m$：训练样本数量
- $x^{(i)}$：第 i 个输入数据
- $y^{(i)}$：第 i 个真实结果
- $\hat{y}^{(i)}$：第 i 个预测结果
- $f_{w,b}(x^{(i)})$：模型对第 i 个样本的预测值
- $\sum$：把所有样本的误差加起来

---

### 6. 为什么要平方？

误差是：$\hat{y}$ - y

但是误差可能是正数，也可能是负数，所以使用平方：($\hat{y}$ - y)^2

作用是：
1. 让误差都变成正数
2. 让大的误差受到更大惩罚
3. 方便后面做梯度下降

---

### 7. 为什么要除以 m？

因为我们有 $m$ 个训练样本。

除以 $m$ 的意思是：求平均误差。

这样无论训练数据有多少个，都可以用一个平均 cost 来衡量模型好坏。

---

### 8. 为什么前面有 1/2？

公式里有：
$$ \frac{1}{2m}$$
这里的 2 主要是为了后面求导更方便。

目前可以先记住：

> $\frac{1}{2}$ 不影响我们理解 cost 的大小，只是数学上方便后面计算梯度。

---

### 9. 总结

| 内容 | 公式 / 说明 |
|---|---|
| model | $\hat{y}=f_{w,b}(x)=wx+b$ |
| parameters | $w,b$ |
| Cost Function | $J(w,b)=\frac{1}{2m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)})-y^{(i)})^2$ |
| Objective | $\min_{w,b} J(w,b)$ |


---

### 10. 这一节我需要记住什么？

固定要记住：

1. 成本函数用来衡量模型预测得好不好
2. $J(w,b)$ 表示当前 $w$ 和 $b$ 对应的成本
3. cost 越小，模型越好
4. 线性回归的目标，就是找到让 $J(w,b)$ 最小的参数 $w,b$。
5. 误差 = 预测值 - 真实值

---

> 成本函数告诉我们模型好不好，梯度下降会帮助我们自动找到更好的 w 和 b。

```python
"""
2,准备训练数据
    这里使用课程中的两个房价数据点

    x_train：房子面积
    y_train：房价
    例如：
        x = 1.0 表示 1000 平方英尺
        y = 300 表示 300,000 美元
"""

# 创建输入数据
x_train = np.array([1.0, 2.0])

# 创建真实结果
y_train = np.array([300.0, 500.0])

# 打印输入数据
print("x_train:", x_train)

# 打印真实结果
print("y_train:", y_train)
```

```python
"""
3,实现成本函数 compute_cost
    成本函数用来衡量模型预测值和真实值之间的差距
    公式：
        J(w,b)=1/(2m) * sum((f_wb - y)^2)
    其中：
        f_wb = w * x + b
    目标：
        找到让 J(w,b) 最小的 w 和 b
"""

# 定义成本函数
# 参数：x 输入数据， y 真实结果， w 权重，b 偏置
def compute_cost(x, y, w, b):

    m = x.shape[0]  # 获取训练样本数量
    cost_sum = 0 # 创建变量 cost_sum，用来累加所有样本的误差平方

    # 使用 for 循环，依次处理每一个训练样本
    for i in range(m):

        # 计算第 i 个样本的预测值
        # 线性回归模型：f_wb = w * x + b
        f_wb = w * x[i] + b

        # 计算预测值和真实值之间的误差平方
        # 误差 = 预测值 - 真实值
        cost = (f_wb - y[i]) ** 2

        # 把当前样本的误差平方累加到 cost_sum 中
        cost_sum = cost_sum + cost

    # 根据成本函数公式计算最终成本
    # J(w,b) = 1/(2m) * sum(error^2)
    total_cost = (1 / (2 * m)) * cost_sum

    # 返回最终成本
    return total_cost
```

```python
"""
4,测试成本函数
    用不同的 w,b 测试 cost
    如果 w,b 选得好，cost 应该很小
    如果 w,b 选得差，cost 应该很大
"""

# 第一组参数：比较好的参数
# 对于当前数据，w=200,b=100 可以刚好拟合两个点
w_good = 200
b_good = 100

# 计算好参数对应的成本
cost_good = compute_cost(x_train, y_train, w_good, b_good)

# 打印结果
print("w = 200, b = 100 时的 cost:", cost_good)


# 第二组参数：比较差的参数
# w=0,b=0 时，所有预测值都是 0
w_bad = 0
b_bad = 0

# 计算差参数对应的成本
cost_bad = compute_cost(x_train, y_train, w_bad, b_bad)

# 打印结果
print("w = 0, b = 0 时的 cost:", cost_bad)
```

## 6. 梯度下降(算法) Gradient Descent 10-15
前面已经知道：

- 模型：$f_{w,b}(x)=wx+b$
- (策略)成本函数：$ J(w,b) = \frac{1}{2m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)}) - y^{(i)})^2 $
- (算法)梯度下降是目标：找到让 $J(w,b)$ 最小的 $w,b$


---
### 1. 梯度下降

梯度下降的作用：

> 不断调整 $w,b$，让成本函数 $J(w,b)$ 变小。

可以理解成：

- 现在站在山坡上，360度旋转，为了下山，应该朝那个方向迈步
- 找到下坡方向
- 往下坡方向走一步
- 重复很多次
- 最后接近最低点

最低点就是：

$J(w,b)$ 最小的位置。

核心：

> 梯度下降就是一步一步走向成本函数的最低点。

---

### 2. 实现梯度下降 

梯度下降会不断更新参数：

| 参数 | 更新公式 |
|---|---|
| $w$ | $w = w - \alpha \frac{\partial J(w,b)}{\partial w}$ |
| $b$ | $b = b - \alpha \frac{\partial J(w,b)}{\partial b}$ |

其中：

- = 是给w赋值
- $\alpha$：学习率（learning rate）通常是0-1之间的小正数，控制下坡时步子的大小
- 模型：$ f_{w,b}(x^{(i)})=wx^{(i)}+b $
- 成本函数：$ J(w,b) = \frac{1}{2m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)}) - y^{(i)})^2 $
- $ \frac{\partial J(w,b)}{\partial w}$：告诉 $w$ 应该怎么变，求导后：$\frac{1}{m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)}) - y^{(i)})*x^{(i)}$
- $ \frac{\partial J(w,b)}{\partial b}$：告诉 $b$ 应该怎么变，求导后：$\frac{1}{m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)}) - y^{(i)})$

重点：

> $w$ 和 $b$ 要同时更新。

---

### 3. 梯度下降直觉 

导数可以理解成：

> 当前点的斜率。

用开口向上的抛物线举例（U）假设b为0

如果某个点在极小值的右侧，则(导数)斜率是正的：w = w-求导后的结果，最终会得到一个更小的w

- 在图上表示参数应该往左走，减小w的值
- 让 代价cost J(w)变小

如果在极小值的左侧，则(导数)斜率是负的：w = w-求导后的结果，最终会得到一个更大的w

- 在图上表示参数应该往右走，增加w的值。
- 让代价cost J(w)变小

所以梯度下降会根据斜率判断：
> 参数应该往哪个方向调整。
> 
> 核心：斜率决定方向，学习率决定步子大小。

---

### 4. 学习率

学习率 $\alpha$ 控制每次走多大一步。

如果 $\alpha$ 太小：每一步很小，下降很慢，需要训练很多次

如果 $\alpha$ 太大：每一步太大，可能跨过最低点，使cost 可能越来越大

所以：

> 学习率不能太大，也不能太小。
> 
> 当接近局部最小值时，导数会越来越小，梯度下降会采取更小的步骤(学习率a)，也就是学习率逐渐减小
---

### 5. 线性回归的梯度下降

当使用线性回归的平方误差成本函数后，成本函数 $J(w,b)$ 不会有多个局部最小值，是一个凸函数。(从上往下看)

也就是：

> 像一个碗一样，只有一个最低点。

所以线性回归中使用梯度下降时：

- 不用担心卡在很多局部最低点
- 目标就是找到这个碗底
- 碗底对应的 $w,b$ 就是最优参数

核心：

> 对线性回归来说，梯度下降最终会朝着全局最低点前进。

---

### 6. 运行梯度下降

运行梯度下降时，要观察：

- cost 是否越来越小
- $w,b$ 是否逐渐接近合适的值
- 预测线是否越来越贴近真实数据

如果 cost 一直下降：

> 说明梯度下降在正常工作。

如果 cost 越来越大：

> 可能是学习率太大。

---

### 梯度下降总结

本部分需要记住：

1. 成本函数 $J(w,b)$ 用来衡量模型好不好
2. 梯度下降用来自动调整 $w,b$
3. 目标是让 $J(w,b)$ 越来越小
4. 学习率 $\alpha$ 决定每次走多大一步
5. 线性回归的成本函数像一个碗，最低点就是最优参数

最核心一句话：

> 梯度下降就是不断更新 $w,b$，让成本函数 $J(w,b)$ 一步一步变小。
