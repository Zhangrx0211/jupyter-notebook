# C1-W2 多变量线性回归学习总览

## 1. 这一章要学什么？

C1-W2 主要学习：

- 多变量线性回归
- 向量化
- 多变量梯度下降
- 特征缩放
- 学习率选择
- 特征工程
- scikit-learn

这一章是在 C1-W1 一元线性回归的基础上继续扩展。

C1-W1 中模型只有一个输入特征：

$f_{w,b}(x)=wx+b$

C1-W2 中模型有多个输入特征：

$f_{\vec{w},b}(\vec{x})=w_1x_1+w_2x_2+...+w_nx_n+b$

也可以写成：

$f_{\vec{w},b}(\vec{x})=\vec{w}\cdot\vec{x}+b$

核心变化：

> 从一个特征 $x$，变成多个特征 $\vec{x}$。
