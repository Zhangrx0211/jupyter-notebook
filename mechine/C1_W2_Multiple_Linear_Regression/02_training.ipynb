## 4.多变量梯度下降、特征缩放与学习率

### 这一节要解决什么问题？
有了多个特征之后，模型参数也变多了。

C1-W1 只需要更新：

- $w$
- $b$

C1-W2 需要更新：

- $w_1,w_2,w_3,\cdots,w_n$
- $b$

所以这一部分要解决的问题是：

> 如何训练多变量线性回归模型，并让梯度下降更快、更稳定？

---

### 核心公式 / 核心概念

多变量线性回归模型：

$$
f_{\vec{w},b}(\vec{x})=\vec{w}\cdot\vec{x}+b
$$

成本函数仍然是：

$$
J(\vec{w},b)=\frac{1}{2m}\sum_{i=0}^{m-1}(f_{\vec{w},b}(\vec{x}^{(i)})-y^{(i)})^2
$$

梯度下降仍然是不断更新参数：

$$
w_j = w_j - \alpha \frac{\partial J(\vec{w},b)}{\partial w_j}
$$

$$
b = b - \alpha \frac{\partial J(\vec{w},b)}{\partial b}
$$

特征缩放的作用：

> 让不同特征的数值范围更接近，使梯度下降更快收敛。

学习率 $\alpha$ 的作用：

> 控制每次更新参数时走多大一步。

---

### 这节我需要记住什么？

- 多变量线性回归中，$w$ 从一个数字变成一组参数
- 梯度下降的核心思想不变，仍然是让 $J(\vec{w},b)$ 变小
- 特征数值范围差太多，会让梯度下降变慢
- 特征缩放可以让训练更快、更稳定
- 学习率太小，训练很慢
- 学习率太大，可能导致 cost 发散
- 判断梯度下降是否正常，主要看 cost 是否逐渐下降

最核心：

> 多变量训练的目标仍然是让成本函数越来越小。

---

### 和前面内容的关系

C1-W1 的梯度下降只更新一个 $w$：

$$
w = w - \alpha dj\_dw
$$

C1-W2 的梯度下降要更新多个 $w$：

$$
w_1,w_2,\cdots,w_n
$$

但本质没变：

> 都是通过不断更新参数，让成本函数 $J$ 下降。
