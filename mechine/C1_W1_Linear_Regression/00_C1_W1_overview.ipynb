## 2. 总体学习流程
C1-W1 的整体流程可以理解为：

数据 → 模型 → 预测 → 误差 → 成本函数 → 梯度下降 → 更好的模型

也就是：

1. 给程序一些训练数据
2. 选择一个线性回归模型
3. 用模型做预测
4. 比较预测值和真实值的差距
5. 用成本函数衡量误差大小
6. 用梯度下降自动调整参数
7. 得到更好的预测模型

## 3. 第一部分：线性回归模型
对应文件：

01_model_representation.ipynb

### 这一部分要解决什么？

给一些训练数据，例如： x 表示输入数据，y 表示真实结果。

>程序要学习出一条直线，用这条直线进行预测。

### 线性回归模型的作用

线性回归模型的作用是：

> 用一条直线拟合数据，并根据输入 $x$ 预测输出 $\hat{y}$。
> 
> 给模型一个 $x$，模型就会根据 $w,b$ 计算出预测值。

公式：$ \hat{y}=f_{w,b}(x)=wx+b $

### 符号含义

| 符号 | 含义 |
|---|---|
| $x$ | 输入数据 |
| $y$ | 真实值 |
| $\hat{y}$ | 预测值 |
| $w$ | 权重，控制直线的倾斜程度 |
| $b$ | 偏置，控制直线的上下位置 |
| $f_{w,b}(x)$ | 模型预测函数 |

### 这一部分要学会什么？

- 创建训练数据
- 画出训练数据
- 理解模型公式
- 实现模型预测
- 画出预测直线
- 明白不同 $w,b$ 会得到不同的预测直线
---



## 4. 第二部分：成本函数
对应文件：

02_cost_function.ipynb

### 这一部分要解决什么？

有了模型之后，需要判断：

> 当前这组 $w,b$ 好不好？

成本函数就是用来衡量模型预测误差的。

### 成本函数的作用

- cost 越小，说明模型预测越好
- cost 越大，说明模型预测越差
>目标是找到让 cost 最小的 $w,b$
>
>成本函数用一个数字衡量模型预测得好不好。

公式：$ J(w,b)=\frac{1}{2m}\sum_{i=0}^{m-1}(f_{w,b}(x^{(i)})-y^{(i)})^2 $

### 符号含义

| 符号 | 含义 |
|---|---|
| $J(w,b)$ | 成本函数 |
| $m$ | 训练样本数量 |
| $f_{w,b}(x^{(i)})$ | 第 i 个样本的预测值 |
| $y^{(i)}$ | 第 i 个样本的真实值 |
| $f_{w,b}(x^{(i)})-y^{(i)}$ | 预测误差 |

### 这一部分要学会什么？

- 理解 $J(w,b)$ 用来衡量预测误差
- 看懂不同 $w,b$ 会对应不同 cost
- 明白 cost 越小，模型越好
- 看懂左图中预测线和真实数据之间的误差
- 看懂右图中不同参数对应的成本
- 看懂 3D 图中的碗底就是 $J(w,b)$ 最小的位置
- 明白目标是找到让 $J(w,b)$ 最小的 $w,b$
---




## 5. 第三部分：梯度下降
对应文件：

03_gradient_descent.ipynb

### 这一部分要解决什么？

成本函数告诉我们模型好不好，但还没有告诉我们：

> 怎么自动找到最好的 $w,b$？

梯度下降就是用来自动调整参数的方法。

### 梯度下降的作用

梯度下降可以理解成：

- 站在山坡上
- 找到下坡方向
- 往下坡方向走一步
- 重复很多次
- 最后接近最低点
- 
最低点就是成本函数最小的位置。

更新公式
| 参数 | 更新公式 |
|---|---|
| $w$ | $w = w - \alpha \frac{\partial J(w,b)}{\partial w}$ |
| $b$ | $b = b - \alpha \frac{\partial J(w,b)}{\partial b}$ |

### 学习率

$\alpha$ 是学习率，控制每一步走多大。

如果学习率太小：下降很慢，训练时间长

如果学习率太大： 可能跨过最低点，cost 可能越来越大

### 这一部分要学会什么？

- 理解梯度下降是用来自动更新参数的
- 理解学习率 $\alpha$ 的作用
- 理解 $w,b$ 要同时更新
- 看懂 cost 是否在逐渐下降
- 明白梯度下降的目标是让 $J(w,b)$ 越来越小
- 最终学会实现 `compute_gradient()` 和 `gradient_descent()`

### 核心记忆

> 梯度下降就是不断更新 $w,b$，让 $J(w,b)$ 一步一步变小。

## 7. 最核心的一句话

> 线性回归的训练过程，就是不断调整 $w,b$，让预测值更接近真实值，让成本函数 $J(w,b)$ 变得最小。
