

### 01_model_representation.ipynb

需要实现一个简单的一元线型回归模型:$f_{w,b}(x) = wx + b$

我们会给程序一些训练数据，例如：

- x 表示输入数据，比如房子面积，y 表示真实结果，比如房价

程序要学习出一条直线：
$f_{w,b}(x) = wx + b$
通过不断调整参数 w 和 b，让预测值越来越接近真实值。


代码流程：
1. 导入工具包
2. 创建训练数据
3. 画出训练数据
4. 实现线性回归模型
5. 测试模型预测结果
6. 画出模型预测直线
---



```python
""" 
1,导入工具包
    import numpy as np
    import matplotlib.pyplot as plt
"""
# 导入 numpy，并且给它起一个简短名字 np
# numpy 主要用来处理数字、数组、矩阵
import numpy as np

# 导入 matplotlib 里的 pyplot，并且给它起一个简短名字 plt
# pyplot 主要用来画图，比如散点图、折线图
import matplotlib.pyplot as plt


print("环境测试成功") # 打印一行文字，用来确认当前代码可以正常运行
print(np.__version__) # 打印 numpy 的版本号，确认 numpy 已经成功导入


"""
2,创建训练数据
    np.array([...])：创建一个 numpy 数组。
"""
# 创建 x_train，用来保存输入数据
x_train = np.array([1.0, 2.0]) # 这里的 x 可以理解为房子大小

# 创建 y_train，用来保存真实结果
y_train = np.array([300.0, 500.0]) # 这里的 y 可以理解为房价

print("x_train:", x_train) # 打印 x_train，查看输入数据
print("y_train:", y_train) 

# 打印训练样本数量
# x_train.shape[0] 表示 x_train 里面有几个数据
print("训练样本数量:", x_train.shape[0])


"""
3,画出训练数据
    plt.scatter(x_train, y_train, marker='x')
    plt.title("标题")
    plt.xlabel("X 名字")
    plt.ylabel("Y 名字")
    plt.show()
"""
# 使用 scatter 函数画散点图
# x_train 是横坐标数据，y_train 是纵坐标数据
# marker='x' 表示用 x 形状标记每个点
plt.scatter(x_train, y_train, marker='x')

plt.title("Housing Prices") # 设置图表标题
plt.xlabel("Size") # 设置横坐标名称
plt.ylabel("Price") # 设置纵坐标名称
plt.show() # 显示图像

```

```python
"""
4,实现线性回归模型
    线性回归模型的公式是:f_wb = w * x + b
    x 是输入数据,w 是权重:决定直线的倾斜程度,b 是偏置:决定直线上下移动的位置 
    f_wb 是模型预测出来的结果

    m = x_train.shape[0]：获取训练样本数量。 
    np.zeros(m) 表示创建 m 个 0    
    for i in range(m):  如果 m = 2,那么 i 会依次等于 0、1
"""
# 定义一个函数，用来计算线性回归模型的预测值
# x 表示输入数据，w 表示权重，b 表示偏置，是这个函数需要的三个输入
def compute_model_output(x, w, b):

    # 获取训练数据的数量
    m = x.shape[0]  # 比如 x_train = [1.0, 2.0]，那么 m = 2

    # 创建一个全是 0 的数组，用来保存每一个 x 对应的预测结果
    f_wb = np.zeros(m) #如果 m = 2，那么 f_wb 一开始就是 [0.0, 0.0]

    # 使用 for 循环，依次计算每一个 x 的预测值
    # range(m) 表示生成 0 到 m-1 的数字
    for i in range(m):  # 如果 m = 2，那么 i 会依次等于 0、1

        # 使用线性回归公式计算预测值
        # x[i] 表示第 i 个输入数据
        f_wb[i] = w * x[i] + b
    return f_wb # 返回所有预测结果


"""
5,测试线性回归模型
    手动设置 w 和 b 的值，看看模型预测的结果和真实结果的差距
"""
# 手动设置 w和b 的值
w = 200
b = 100

# 调用 compute_model_output 函数，计算预测结果
# x_train 是输入数据
tmp_f_wb = compute_model_output(x_train, w, b)
"""
例:x_train = np.array([1.0, 2.0]), w = 200, b = 100, 模型公式：预测值 = 200 * x + 100
当x=1.0时，预测值=300。当x=2.0时，预测值=500。
"""
print("模型预测值:", tmp_f_wb) # 打印预测结果


"""
6,画出模型预测直线
    plt.scatter(x_train, y_train)：画真实训练数据的 散点图
    plt.plot(x_train, tmp_f_wb)：画模型预测出来的   直线
    如果预测直线刚好穿过真实数据点，说明当前 w 和 b 拟合得很好
"""
# 使用 scatter 函数画出真实训练数据的散点图
# label="Actual Value" 表示散点图的名字 Actual Value
plt.scatter(x_train, y_train, marker='x', label="Actual Value")

# 使用 plot 函数画出模型预测出来的直线
plt.plot(x_train, tmp_f_wb, label="Prediction")

plt.title("Linear Regression Model") 
plt.xlabel("Size") 
plt.ylabel("Price") 
plt.show()
plt.legend() # 显示散点的名字和直线的名字
```

```python

"""
7,用当前模型做一次预测
    前面我们已经手动设置了：
        w = 200
        b = 100

    现在输入一个新的 x
    用公式 y_hat = w * x + b
    预测对应的 y

"""

# 设置一个新的输入数据
# 这里 x_i = 1.2，可以理解为房子大小是 1.2
x_i = 1.2

# 使用线性回归公式进行预测
prediction = w * x_i + b

# 打印预测结果
print(f"${prediction:.0f} thousand dollars")
```
